## Style and Color

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
```

## Spine Removal

```python
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='seismic')
```

## Size and Aspect

You can use matplotlib's*plt.figure(figsize=(weight,height) * to change the size of most seaborn plots.
You can control the size and aspect ratio of most seaborn grid plots by passing in parameters: size, and aspect. For example:

```python
# Non Grid Plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Grid Type Plot
sns.lmplot(x='total_bill',y='tip',size=2,aspect=4,data=tips)
```
