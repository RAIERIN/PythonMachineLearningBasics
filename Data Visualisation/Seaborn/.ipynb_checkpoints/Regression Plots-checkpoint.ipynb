## Regression Plots
Seaborn has many built-in capabilities for regression plots, however we won't really discuss regression until the machine learning section of course, so we will only cover the Implot() function for now.

Implot allows you to display linear models, but it also conveniently allows you to split up those plots based off features, as well as coloring the hue based off of features:

```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Working with Markers

Implot kwargs ge4t passed through to regplot which is a more general form of Implot(). regplot has a scatter_kws parameter that gets passed to plt.scatter. So you want to set the s parameter in the dictionary, which corresponds (a bit confusingly) to the squared markersize. In other words you end up passing a dictionary with the base matplotlib arguements, in this case, s for size of a scatter plot. 

```python


sns.lmplot(x='total_bill',y="tip",data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':100})
```

## Using a Grid
We can add more variable seperation through columns and rows with the use of a grid. Just indicate this with the col or row arguements:

```python
sns.lmplot(x='total_bill',y="tip",data=tips,col='day',hue='sex',aspect=0.6,height=8)
```

## Aspect and Size
Seaborn figures can have their size and aspect ratio adjusted with the size and aspect parameters:

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm',
          aspect=0.6,size=8)
```
